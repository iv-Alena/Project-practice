```python
from inference_sdk import InferenceHTTPClient

CLIENT = InferenceHTTPClient(
    api_url="https://serverless.roboflow.com",
    api_key="VDdoyat8yYSu3wKgBWPA"
)
```

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

def print_pic(sample, result):
  img_path = sample
  img = Image.open(img_path).convert("RGB")
  draw = ImageDraw.Draw(img)
  preds = result.get("predictions", [])

  for p in preds:
      # координаты центра и ширины/высоты из Roboflow
      x, y = p["x"], p["y"]
      w, h = p["width"], p["height"]
      cls_name = p.get("class", "object")
      conf = p.get("confidence", 0)
      x0 = x - w / 2
      y0 = y - h / 2
      x1 = x + w / 2
      y1 = y + h / 2

      # Рамкf
      draw.rectangle([x0, y0, x1, y1], outline="red", width=3)

      # Подпись (класс + вероятность)
      text = f"{cls_name} ({conf:.2f})"
      draw.text((x0, y0 - 10), text, fill="red")


  plt.figure(figsize=(4, 4))
  plt.imshow(img)
  plt.axis("off")
  plt.show()
```

```python
result1 = CLIENT.infer('sample1.jpg', model_id="ne-bird/7")
print('Класс распознанной пицы:', result1['predictions'][0]['class'])
print_pic('sample1.jpg', result1)
```
